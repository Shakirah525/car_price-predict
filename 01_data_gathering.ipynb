## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('Libraries loaded successfully!')
```

## 2. Load Dataset

```python
# Load the dataset
df = pd.read_csv('../data/CarPrice_Assignment.csv')
print(f'Dataset loaded successfully!')
print(f'Shape: {df.shape[0]} rows x {df.shape[1]} columns')
```

## 5. Missing Values Analysis

```python
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0]

if missing_df.empty:
    print('No missing values found in the dataset!')
else:
    print('Columns with missing values:')
    print(missing_df.sort_values('Missing %', ascending=False))
```

## 6. Target Variable Analysis: Price Distribution

```python
print('='*60)
print('PRICE STATISTICS')
print('='*60)
print(df['price'].describe())
print(f'\nSkewness : {df["price"].skew():.3f}')
print(f'Kurtosis : {df["price"].kurtosis():.3f}')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Histogram
axes[0].hist(df['price'], bins=30, color='steelblue', edgecolor='white')
axes[0].set_title('Price Distribution', fontsize=14)
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Frequency')

# Boxplot
axes[1].boxplot(df['price'], patch_artist=True, boxprops=dict(facecolor='steelblue', alpha=0.7))
axes[1].set_title('Price Boxplot', fontsize=14)
axes[1].set_ylabel('Price ($)')

# Log-transformed
axes[2].hist(np.log1p(df['price']), bins=30, color='coral', edgecolor='white')
axes[2].set_title('Log-Transformed Price', fontsize=14)
axes[2].set_xlabel('Log(Price)')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.savefig('../reports/01_price_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('Plot saved to reports/')
```

## 7. Column Overview & Data Types Classification

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print(f'Numerical Columns ({len(numerical_cols)}): {numerical_cols}')
print(f'\nCategorical Columns ({len(categorical_cols)}): {categorical_cols}')
```

```python
# Unique value counts for categorical columns
print('='*60)
print('UNIQUE VALUES IN CATEGORICAL COLUMNS')
print('='*60)
for col in categorical_cols:
    print(f'\n{col} ({df[col].nunique()} unique):')
    print(df[col].value_counts().to_string())
```

## 8. Save Structure Info

```python
# Save data info summary
summary = {
    'total_rows': df.shape[0],
    'total_columns': df.shape[1],
    'numerical_cols': numerical_cols,
    'categorical_cols': categorical_cols,
    'missing_values': df.isnull().sum().to_dict(),
    'price_stats': df['price'].describe().to_dict()
}

import json
with open('../reports/01_data_summary.json', 'w') as f:
    json.dump(summary, f, indent=2)

# Save raw copy
df.to_csv('../data/01_raw_snapshot.csv', index=False)
print('Data summary saved to reports/01_data_summary.json')
print('Raw snapshot saved to data/01_raw_snapshot.csv')
print('\nNotebook 1 Complete!')
```
