```python
import pandas as pd
```

```python
# Load Data
df = spark.table("car_price_gathering_temp")
df_pd = df.toPandas()
print(f"Dataset shape: {df_pd.shape}")
```

```python
#3: Check Missing Values
missing_values = df_pd.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
#4: Data Overview
display(df_pd.describe())
# Shows statistics and categorical value counts
```

```python
#5: Check and Remove Duplicates
duplicates = df_pd.duplicated().sum()
if duplicates > 0:
    df_pd = df_pd.drop_duplicates()
```

```python
#6: Save Cleaned Data
df_cleaned = spark.createDataFrame(df_pd)
df_cleaned.write.mode("overwrite").saveAsTable("car_price_cleaned")
```
